# Aviation Data Analysis.

The goal of the analysis done in this notebook is to help the company expand into new industries to diversify its portfolio by entering the aviation sector by  purchasing and operating airplanes and for both commercial and private enterprises.

To accomplish this, it's important to identify:
1. Aircraft with the lowest risk profiles based on historical data.
2. How operational and environmental factors such as weather conditions and flight phases contribute to aviation risks.

The result of this analysis should give the company actionable insights to guide purchasing decisions and risk mitigation strategies.

### 1. Importing the relevant libraries and taking a look at the data.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Setting low_memory = False to avoid DtypeWarning
df = pd.read_csv('AviationData.csv', encoding = 'latin-1', low_memory = False)
```

### 2. Data cleaning.

To deal with missing data, the dataframe was modified to only contain the relevant columns.

```python
df = df[['Make','Model','Aircraft.damage','Aircraft.Category','Total.Fatal.Injuries','Total.Serious.Injuries','Total.Minor.Injuries','Total.Uninjured','Weather.Condition','Broad.phase.of.flight','Purpose.of.flight','Event.Date']]
```

```python
df.isna().sum()
```

```python
# Dropping rows with missing values in the Make, Model and Total Fatal injuries columns

df = df.dropna(subset = ['Make','Model','Total.Fatal.Injuries'])
```

```python
# Replacing rows with missing values in the Total.Serious.Injuries column with the median.

df['Total.Serious.Injuries'].fillna(df['Total.Serious.Injuries'].median(), inplace = True)
```

```python
df['Total.Minor.Injuries'].fillna(0, inplace = True)
```

```python
df['Weather.Condition'].fillna(df['Weather.Condition'].mode(), inplace = True)
```

```python
df['Purpose.of.flight'].fillna('unknown', inplace = True)
```

```python
df.info()
```

### 3. Analyzing the Data.

To begin by analyzing risk metrics such as Total.Fatal.Injuries, Total.Serious.Injuries, and Total number of Incidents.

```python
# To create a column that will help count the total number of incidents.
df['Total.Incidents'] = 1
```

```python
# To know the Fatality rate for each incident.
df['Fatality.Rate'] = df['Total.Fatal.Injuries']/df['Total.Incidents']

# The same for Serious injury rate.
df['Serious.Injury.Rate'] = df['Total.Serious.Injuries']/df['Total.Incidents']
```

```python
df['Serious.Injury.Rate'].value_counts()
```

```python
# To create a dataframe containing the risk metrics.
risk_summary = df.groupby(['Make', 'Model', 'Fatality.Rate']).agg({
    'Total.Incidents': 'sum',
    'Total.Fatal.Injuries': 'sum',
    'Total.Serious.Injuries': 'sum'
}).reset_index()
```

To remove aircraft with very few incidents which might be misleading due to insufficient data. Setting the minimum number of incidents for each aircraft to be analyzed to 100 incidents.

```python
risk_summary = risk_summary[risk_summary['Total.Incidents'] >= 100]
```

```python
risk_summary
```

```python
# To create a dataframe of aircraft with the lowest fatality rates.

prioritized_aircraft = risk_summary.sort_values(by='Fatality.Rate', ascending = True)
```

```python
prioritized_aircraft
```

```python
top_aircraft = prioritized_aircraft.head(10)
```

```python
top_aircraft
```

Since the safest aircraft in the data all have a Fatality rate of 0, it is better to use another risk metric to determine the risk level of each aircraft. The chosen metric was 'Total.Serious.Injuries'

```python
 safe_aircraft = top_aircraft.sort_values(by='Total.Serious.Injuries', ascending = True)
```

```python
safe_aircraft.head(10)
```

Next is to plot charts of the safest and riskiest aircraft models as per the Total.Serious.Injuries risk metric.

```python
safe_aircraft.head()
```

```python
safe_aircraft['Make_Model'] = safe_aircraft['Make'] + ' ' + safe_aircraft['Model']

top_safest = safe_aircraft.head(5)

plt.figure(figsize=(12, 6))
plt.bar(top_safest['Make_Model'], top_safest['Total.Serious.Injuries'], color = 'skyblue')
plt.title('Top 5 Aircraft with Lowest Total Serious Injuries Rates')
plt.xlabel('Aircraft Model')
plt.ylabel('Total Serious Injuries')
plt.show()
```

The above are the five safest aircraft based on Total Serious Injuries Rates

```python
Riskiest_aircraft = risk_summary.sort_values(by='Total.Serious.Injuries', ascending = False)
```

```python
Riskiest_aircraft
```

```python
Riskiest_aircraft['Make_Model'] = Riskiest_aircraft['Make'] + ' ' + Riskiest_aircraft['Model']

top_riskiest = Riskiest_aircraft.head(5)

plt.figure(figsize=(12, 6))
plt.bar(top_riskiest['Make_Model'], top_riskiest['Total.Serious.Injuries'], color='skyblue', alpha=0.7)

plt.title('Top 5 Riskiest Aircraft by Total Serious Injuries', fontsize=16)
plt.xlabel('Aircraft (Make & Model)', fontsize=14)
plt.ylabel('Total Serious Injuries', fontsize=14)

plt.xticks(rotation=45, ha='right', fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.6) # Adding grid lines for better readability

plt.tight_layout()
plt.show()
```

The above are the 5 riskiest aircrafts based on Total Serious Injuries Rates

### Analyzing Operational and Environmental Variables.

These included Weather.Condition, Purpose.of.flight, and Phase.of.flight

```python
weather_incidents = df.groupby('Weather.Condition')['Total.Incidents'].sum().reset_index()
weather_incidents.columns = ['Weather Condition', 'Incident Count']
weather_incidents = weather_incidents.sort_values(by = 'Incident Count', ascending = False)

weather_incidents
```

Most of the incidents happened during 'VMC' weather but this doesn't tell us much since these are suitable conditions in aviation and is when most aircraft are operated.

Now to look at the total incidents for every flight purpose.

```python
purpose_incidents = df.groupby('Purpose.of.flight')['Total.Incidents'].sum().reset_index().sort_values(by = 'Total.Incidents', ascending = False)

purpose_incidents
```

```python
plt.figure(figsize=(12, 6))
plt.bar(purpose_incidents['Purpose.of.flight'], purpose_incidents['Total.Incidents'], color='skyblue', edgecolor='black')

plt.title('Total Incidents by Purpose of Flight', fontsize=16)
plt.xlabel('Purpose of Flight', fontsize=14)
plt.ylabel('Total Incidents', fontsize=14)

plt.xticks(rotation=45, ha='right', fontsize=12) # Rotating for readability.

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

The personal Flight purpose has the highest number of incidents and is therefore the riskiest.

Now to analyse risk level at the different phases of flight.

```python
phase_incidents = df.groupby('Broad.phase.of.flight')['Total.Incidents'].sum().reset_index()

phase_incidents.columns = ['Phase of Flight', 'Total Incidents']

phase_incidents = phase_incidents.sort_values(by='Total Incidents', ascending=False)
```

```python
phase_incidents
```

To plot a chart representing this:

```python
plt.figure(figsize=(12, 6))
plt.bar(phase_incidents['Phase of Flight'], phase_incidents['Total Incidents'], color='skyblue', edgecolor='black')

plt.title('Total Incidents by Phase of Flight', fontsize=16)
plt.xlabel('Phase of Flight', fontsize=14)
plt.ylabel('Total Incidents', fontsize=14)

plt.xticks(rotation=45, ha='right', fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

Landing is the riskiest phase of flight followed by Takeoff, cruise, and maneuvering.

```python
df['Year'] = pd.to_datetime(df['Event.Date']).dt.year
annual_trends = df.groupby('Year').agg({'Total.Fatal.Injuries': 'sum'}).reset_index()
```

To analyze the trend in fatality over the course of time:

```python
annual_trends
```

To plot this for data for the past 20 years:

```python
latest_year = annual_trends['Year'].max()
start_year = latest_year - 20
last_20_years = annual_trends[annual_trends['Year'] >= start_year]

serious_injuries_trend = last_20_years.groupby('Year')['Total.Fatal.Injuries'].sum().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(serious_injuries_trend['Year'], serious_injuries_trend['Total.Fatal.Injuries'], marker='o', color='orange')

plt.title('Trend in Total Fatal Injuries Over the Past 20 Years', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Total fatal Injuries', fontsize=14)

plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

Total fatal injuries per year are currently at one of the lowest points for the last 20 years.
